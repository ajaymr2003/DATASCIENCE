```python
import matplotlib.pyplot as plt
import numpy as np

# First graph
x = np.array([2, 10])
y = np.array([2, 10])
  
plt.suptitle('GRAPH')       

plt.subplot(2, 3, 1)       
plt.title('Title 1')
plt.plot(x, y, 'o')

# Second graph (on the right side of the first one)
x = np.array([2, 7, 2,7])
y = np.array([7, 2, 7,2])
plt.subplot(2, 3, 2)         # Second subplot

plt.plot(x,lw=20.5,marker='*' ,c='orange')
plt.plot(y,linestyle='dotted')


x = np.array([2, 7, 2,7])
plt.subplot(2, 3, 3)
plt.plot(x,mfc='yellow',mec='black',marker='o')

x=([1,5])
y=([3,10])
my='TITLE2'
plt.subplot(2,3,4)
plt.grid()
plt.plot(x,y)
plt.title(my)


a=([5,8,10,7])
b=([3,6,9,12])
color=([0,10,20,30])
plt.subplot(2,3,5)
plt.scatter(a,b,alpha=0.5,c=color,cmap='viridis')

a=(['A','B','C',])
b=([5,10,8])
oc="red"
plt.subplot(2,3,6)
plt.bar(a,b,color=oc, width=0.5)
plt.suptitle('GRAPH')

plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 20, 20, 25])
labels = ['35','20','20','25']
colo=['red','blue','blue','black']
ex = [0.1, 0, 0, 0]

plt.pie(y, labels=labels,startangle=90,explode=ex,shadow='true',colors=colo)
plt.legend()
plt.show()
```
